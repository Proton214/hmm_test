```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('perf_original_data.csv')
df['total'] = df['power_energy_pkg_'] + df['power_energy_ram_']
df.to_csv('perf_original_data.csv')
```

```python
SP_data = df[df['Application'] == 'SP']
CG_data = df[df['Application'] == 'CG']
FT_data = df[df['Application'] == 'FT']
BT_data = df[df['Application'] == 'BT']
LU_data = df[df['Application'] == 'LU']
EP_data = df[df['Application'] == 'EP']
MG_data = df[df['Application'] == 'MG']
IS_data = df[df['Application'] == 'IS']
```

```python
SP_data.to_csv('perf_SP.csv')
CG_data.to_csv('perf_CG.csv')
FT_data.to_csv('perf_FT.csv')
BT_data.to_csv('perf_BT.csv')
LU_data.to_csv('perf_LU.csv')
EP_data.to_csv('perf_EP.csv')
MG_data.to_csv('perf_MG.csv')
IS_data.to_csv('perf_IS.csv')
```

```python
corr_sp = SP_data.corrwith(SP_data['total'], method='pearson')
corr_cg = CG_data.corrwith(CG_data['total'], method='pearson')
corr_ft = FT_data.corrwith(FT_data['total'], method='pearson')
corr_bt = BT_data.corrwith(BT_data['total'], method='pearson')
corr_lu = LU_data.corrwith(LU_data['total'], method='pearson')
corr_ep = EP_data.corrwith(EP_data['total'], method='pearson')
corr_mg = MG_data.corrwith(MG_data['total'], method='pearson')
corr_is = IS_data.corrwith(IS_data['total'], method='pearson')
```

```python
test = [corr_sp, corr_cg, corr_ft, corr_bt, corr_lu, corr_ep, corr_mg, corr_is]
```

```python
plt.figure(figsize=(20,13))

x_axis_labels = ['branch_misses','cache_misses','cache_references','instructions','cpu_clock','minor_faults'           
,'page_faults','task_clock','L1_icache_loads','LLC_loads','branch_load_misses','iTLB_load_misses','iTLB_loads'              
,'cpu_branch_misses_','cpu_bus_cycles_','cpu_cache_misses_','cpu_cache_references_','cpu_cpu_cycles_'          
,'cpu_instructions_','msr_aperf_','msr_mperf_','msr_tsc_','cpu_ref_cycles_','power_energy_ram_','power_energy_pkg_'       
,'total']
y_axis_labels = ['SP', 'CG', 'FT', 'BT', 'LU', 'EP', 'MG', 'IS']

ax = sns.heatmap(test, xticklabels=x_axis_labels, yticklabels=y_axis_labels, 
            annot=True, fmt = '.3f', linewidths=.5, cmap='Blues')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```
