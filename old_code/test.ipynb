```python
knl_BT['energy_total'] = knl_BT['Perf_node01_power_energy_total_'].diff()
```

```python
knl_BT
```

```python
knl_BT['energy_total'].describe()
```

```python
knl_BT['energy_updown'] = 0
```

```python
knl_BT
```

```python
q = knl_BT['energy_total'].quantile(.5)
```

```python
knl_BT['energy_total'][knl_BT['energy_total'] > q]
```

```python
knl_BT['energy_updown'] = np.where(knl_BT['energy_total'] > q, 'Up', 'Down')
```

```python
knl_BT
```

```python
knl_BT.to_csv("knl_BT.csv")
```

```python
q1 = knl_BT['energy_total'].quantile(.75)
q2 = knl_BT['energy_total'].quantile(.5)
q3 = knl_BT['energy_total'].quantile(.25)
```

```python
knl_BT['energy_updown'] = np.where((knl_BT['energy_updown'] == 'Up') & (knl_BT['energy_total'] > q1), 'HUp', knl_BT['energy_updown'])
```

```python
knl_BT['energy_updown'] = np.where((knl_BT['energy_updown'] == 'Down') & (knl_BT['energy_total'] < q3), 'HDown', knl_BT['energy_updown'])
```

```python
knl_BT.to_csv("knl_BT.csv")
```
